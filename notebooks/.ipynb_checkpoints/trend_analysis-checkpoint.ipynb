# Trend Analysis of Social Media Engagement
This notebook explores trends over time for likes, shares, and comments to understand the peaks and troughs in user engagement on social media platforms.


## Introduction
We'll be pulling data directly from a PostgreSQL database, processing it with Pandas, and visualizing trends using Matplotlib and Seaborn. This analysis helps in identifying the effectiveness of different content types and posting strategies.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings


from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf

sns.set()
warnings.simplefilter(action='ignore', category=FutureWarning)


# Ensure that the plots are displayed inside the notebook
%matplotlib inline
```

## Database Configuration
Below is the setup for the PostgreSQL database connection. We use a custom class `PostgresDatabase` to manage our database operations.


```python
from packages.database_util import PostgresDatabase

# Initialize database connection using credentials
db = PostgresDatabase(
    host='postgres',
    port=5432,
    database='ope_market',
    user='postgres',
    password='postgres'
)
```

## Data Extraction

We extract likes, shares, and comments data for the accounts 'lit' and 'ope' sorted by the creation date of the posts.


```python
query = """
SELECT created_at, likes_count, share_count
FROM public.posts
WHERE created_at IS NOT NULL
ORDER BY created_at;
"""
data = db.fetch_data(query)
```

## Data Preparation

Convert timestamps to datetime objects and resample the data to daily aggregates. This smooths out noise and fills in gaps without data.


```python
data['created_at'] = pd.to_datetime(data['created_at'])
data.set_index('created_at', inplace=True)
data = data.resample('D').sum()  # Resample data daily
```

## Autocorrelation Analysis

Examine the autocorrelation to check if past values influence future values, which can be crucial for forecasting.


```python
from statsmodels.graphics.tsaplots import plot_acf
import matplotlib.pyplot as plt

plot_acf(data['likes_count'], alpha=0.05)  # Includes 95% confidence interval
plt.title('Autocorrelation of Daily Likes')
plt.xlabel('Lags')
plt.ylabel('Autocorrelation')
plt.show()
```

# Model Fitting with Autoregressive (AR) Model

## Model Fitting
This section demonstrates how to fit an Autoregressive (AR) model to the "Daily Likes" data. We choose the order of the AR model based on the significant lags observed in the ACF plot. An AR(2) model is used here as an example, considering significant autocorrelation at the first two lags.


```python
import numpy as np
import pandas as pd
from statsmodels.tsa.ar_model import AutoReg
from statsmodels.tsa.stattools import acf

# Assuming 'data' is your DataFrame and 'likes_count' is the column with the data
# Convert index to datetime if not already
data.index = pd.to_datetime(data.index)

# Choosing the order of the AR model based on ACF plot
# Let's use AR(2) as a starting example, given significant autocorrelation at lag 1 and 2
model = AutoReg(data['likes_count'], lags=2)
model_fitted = model.fit()

# Print the summary of the model's fit
print(model_fitted.summary())
```

# Interpretation of AutoRegressive (AR) Model Results

## Model Summary

The AutoRegressive (AR) model fitted is an AR(2) model, which means it uses the two previous time steps (`L1`, `L2`) of the series to predict the current value. Here are the key aspects of the model fit:

- **Number of Observations**: The model was fitted on 1559 observations of the `likes_count` data.

- **Model Fit Quality**:
  - **Log Likelihood**: The log likelihood of the model is -3934.598, which helps in comparing this model against other models. Higher values are generally better.
  - **AIC (Akaike Information Criterion)**: 7877.197 - A lower AIC suggests a better model. AIC helps in model selection by balancing model complexity and fit.
  - **BIC (Bayesian Information Criterion)**: 7898.599 - Similar to AIC but with a stricter penalty for model complexity. A lower BIC suggests a better model.

## Coefficients

- **Constant (Intercept)**: 1.1211
  - The model estimates a baseline 'likes_count' of approximately 1.1211 when previous likes are zero.
  - This coefficient is significantly different from zero (p < 0.000), suggesting a strong baseline effect in the data.

- **First Lag (likes_count.L1)**: 0.3392
  - This coefficient means that there is a positive relationship between the previous days likes and the current day's likes. Specifically, each additional like from the previous day contributes about 0.339 likes today.
  - The p-value is less than 0.000, indicating that this relationship is statistically significant.

- **Second Lag (likes_count.L2)**: 0.2129
  - Similarly, each additional like two days ago contributes about 0.213 likes today. This effect is smaller than that of the first lag, which is common in AR models as the influence generally diminishes with older lags.
  - This coefficient is also statistically significant (p < 0.000).

## Roots of the Characteristic Equation

- The roots of the AR characteristic equation provide information about the stability and behavior of the dynamic system modeled by the AR process.
  - **Roots**: The model has one root at 1.5124, which is outside the unit circle, indicating a stable process. The other root at -3.1055 also confirms the model's stability as it lies outside the unit circle.

## Conclusion

The AR(2) model shows a statistically significant dependency of `likes_count` on its past values up to two days back. The model fits well according to the AIC and BIC, and the stability of the model is confirmed by the roots of the characteristic equation. These results suggest that past 'likes' are useful predictors of future 'likes', and the model could be utilized for short-term forecasting.


# Residual Analysis

## Analyzing Residuals
After fitting the AR model, it is crucial to analyze the residuals to ensure that they behave like white noise, which suggests that the model has successfully captured the underlying process. This section plots the residuals and computes their autocorrelation to check for any remaining patterns or autocorrelation.


```python
import matplotlib.pyplot as plt

# Get residuals from the model
residuals = model_fitted.resid

# Plot residuals
plt.figure(figsize=(10,4))
plt.plot(residuals)
plt.title('Residuals from AR Model')
plt.xlabel('Date')
plt.ylabel('Residuals')
plt.show()

# Check for autocorrelation in residuals
lag_acf = acf(residuals, fft=False, nlags=20)
plt.figure(figsize=(10,4))
plt.stem(lag_acf)
plt.title('Autocorrelation Function of the Residuals')
plt.xlabel('Lags')
plt.ylabel('Autocorrelation')
plt.axhline(y=0, linestyle='--', color='gray')
plt.axhline(y=-1.96/np.sqrt(len(residuals)), linestyle='--', color='gray')
plt.axhline(y=1.96/np.sqrt(len(residuals)), linestyle='--', color='gray')
plt.show()
```

# Analysis of Residuals from AR Model

## Overview

The residuals from the AR model provide crucial insights into the fit of the model. Analyzing the residuals helps in determining whether the model has successfully captured all the relevant information, or if there are patterns still left to be modeled. The two key aspects to consider are the behavior of the residuals over time and their autocorrelation.

## Residual Plot Analysis

- **Residual Behavior**: The plot of the residuals over time shows fluctuations around zero, which is generally a good sign indicating that the model is not biased. However, there are several noticeable spikes where the residuals are quite large. These spikes suggest occasional large errors, which could be due to outliers in the data or specific events that the model was unable to predict.

- **Consistency and Variance**: The residuals mostly oscillate in a consistent range around zero, which suggests that the model variance is relatively constant. However, the presence of spikes might indicate potential heteroscedasticity, where the variance of the residuals is not constant over time.

## Autocorrelation of Residuals

- **Lag 0**: The autocorrelation at lag 0 is always 1, as each time series is perfectly correlated with itself.

- **Other Lags**: The autocorrelation function (ACF) of the residuals shows that all the lags, except for the immediate first lag, are within the confidence interval (marked by the dashed lines), which suggests that there is no significant autocorrelation in the residuals. This indicates that the model has captured most of the time-dependent structure in the data.

- **First Lag Concern**: The first lag shows a slight correlation that is just on the edge of the confidence interval. This might not be significant, but it's worth investigating further. It could indicate that the AR model might benefit from including one more lag or possibly adjusting the model to account for this correlation.

## Conclusion

The analysis of the residuals indicates that the AR model performs well in capturing the dynamics of the data with some exceptions. The presence of large spikes in the residuals suggests that there might be outliers or non-standard events that the model is not accounting for. The lack of significant autocorrelation in the residuals, except possibly at the first lag, suggests that the model is adequately specifying the dependencies in the data.

### Recommendations

- **Further Investigation**: Investigate the cause of spikes in the residuals to understand whether they are outliers or events that require special modeling.
- **Model Adjustment**: Consider testing an AR model with an additional lag or a different model specification to better handle any autocorrelation at lag one.
- **Variance Stabilization**: If heteroscedasticity is confirmed, consider using methods like weighted least squares or transforming the data to stabilize the variance.

This residual analysis is crucial for verifying the adequacy of the AR model and guiding further improvements to enhance the model's predictive accuracy.


# Cross-Validation of Time Series Model

## Time Series Cross-Validation
This section covers the implementation of time series cross-validation to evaluate the forecasting performance of the Autoregressive (AR) model. Unlike standard cross-validation methods, time series cross-validation ensures that the train-test split respects the temporal order of observations. This method helps us understand how well our model performs on unseen data over multiple splits, using the `TimeSeriesSplit` from `sklearn.model_selection`.


```python
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_squared_error

# TimeSeries cross-validator
tscv = TimeSeriesSplit(n_splits=5)

# Example of manual TimeSeries cross-validation for simplicity
for train_index, test_index in tscv.split(data['likes_count']):
    train, test = data['likes_count'][train_index], data['likes_count'][test_index]
    model = AutoReg(train, lags=2)
    model_fitted = model.fit()
    predictions = model_fitted.predict(start=test.index[0], end=test.index[-1])
    
    mse = mean_squared_error(test, predictions)
    print(f'Test MSE: {mse}')

    # Plotting actual vs predicted for this fold
    plt.figure(figsize=(10,4))
    plt.plot(test.index, test, label='Actual')
    plt.plot(test.index, predictions, label='Predicted')
    plt.legend()
    plt.title('Forecast vs Actuals')
    plt.show()
```

# Analysis of Actual vs. Predicted Values from AR Model

## Overview

This plot provides a visual comparison between the actual observed values of "Daily Likes" and the predictions made by the Autoregressive (AR) model. This could be a sign that the model parameters are not correctly capturing the dynamics of the data, possibly due to underfitting.

- **Model Diagnostics**: Further diagnostics are required to understand why the model is performing poorly. This could include checking the stationarity of the series, re-examining the autocorrelation in the residuals, or considering additional data transformations.
